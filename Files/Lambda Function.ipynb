```python
def count_primes(num):

    if num < 2:
         return 0
#stor our prime numbers
    primes = [2]
#counter going upto input num
    x = 3

    while x <= num:
        #cheak if x is prime
        for y in range (3,x,2):
            
            if x%y == 0:
              
                x += 2
                break
            else:
               
                primes.append(x)
                x += 2
        print(primes)
        return len(primes)
            
            
```

```python
count_primes(30)
```
