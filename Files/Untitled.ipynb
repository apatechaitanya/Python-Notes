```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10)
y = 2*x+5
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,linewidth = 1.0,linestyle = ":",color = 'r',alpha = 0.5,marker = '+')
plt.title('Line Plot Demo')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(['line1'],loc = 'best')
plt.grid(True)

plt.show()
```
